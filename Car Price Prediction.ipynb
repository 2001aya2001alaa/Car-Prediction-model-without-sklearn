```python
# Select features with the highest correlation

features = car_data[['carwidth', 'curbweight', 'enginesize', 'horsepower']]
features
```

```python
# Shuffle data frame using sample function to split the data into train and test

features = features.sample(frac=1)
print(features)

pred = car_data['price'].sample(frac=1)
print(pred)
```

```python
# Normalization Function

def minmax(feature):
    maxi = feature.max()
    mini = feature.min()
    n = feature.shape[0]
    ft = pd.DataFrame(np.zeros(features.shape))
    ft = (feature-mini)/(maxi-mini)
    return ft
    
features['carwidth'] = minmax(features['carwidth'])
features['curbweight'] = minmax(features['curbweight'])
features['enginesize'] = minmax(features['enginesize'])
features['horsepower'] = minmax(features['horsepower'])
```

```python
ratio = 0.8   #80% Train  20% Test
  
total_rows = features.shape[0]
train_size = int(total_rows*ratio)
  
# Split data into test and train
xtrain = features[0:train_size]
xtest = features[train_size:]

ytrain = car_data['price'][0:train_size]
ytest = car_data['price'][train_size:]

print("X Train data:" ,xtrain.shape[0], "     X Test data:",xtest.shape[0])
print("Y Train data:" ,ytrain.shape[0], "     Y Test data:",ytest.shape[0])
```

```python
xtrain.insert(4, 'Ones', 1)
xtrain
```

```python
xtrain = np.array(xtrain)
ytrain = np.array(ytrain).flatten()
theta = np.array([0,0,0,0,0])  #theta0 theta1
```

```python
m = len(xtrain)

def cost(x, y, theta):
    j = np.sum((x.dot(theta)-y)**2)/2*m
    return j
cost(xtrain, ytrain, theta)
```

```python
its = 1000
alpha = 0.0000001

def gradient_des(x,y,theta, alpha,its):
    history = [0]*its
    #err = np.zeros((164,5))
    for i in range(its):
        f_x = x.dot(theta)
        err = f_x - y
        gradient = x.T.dot(err)/m
        theta = theta - (alpha*gradient)
        costhistory = cost(x,y,theta)
        history[i] = costhistory
    return theta, history

(w, h) = gradient_des(xtrain, ytrain, theta, alpha, its)
print(w)
```

```python
plt.plot(h)
```

```python
ypred = np.dot(xtrain, w)

# Mean Squared Error
mse = np.square(np.subtract(ytrain,ypred)).mean()

print(mse)
```
